<h2><b>WHAT IS NAMED ENTITY RECOGNITION (NER)?</b></h2>

<p>

<ul>
<li>Person: John, Ravi</li>
<li>Date: tomorrow, Monday</li>
<li>Time: 3pm, 6pm</li>
<li>Location: Chennai, Delhi</li>
</ul>

<p>
NER converts unstructured text into structured data, making it easier for machines to process.
</p>

<h2><b>WHY IS NER IMPORTANT IN AI ASSISTANTS?</b></h2>

<p>
NER helps AI systems understand user input by extracting key details such as who, when, time, and location.
</p>

<p><b>EXAMPLE:</b></p>

<p>
Input: "Schedule meeting with John tomorrow at 3pm"
</p>

<ul>
<li>Person → John</li>

<h1><b>STEP 2: RULE-BASED ENTITY EXTRACTION</b></h1>

<h2><b>OVERVIEW</b></h2>

<p>
Rule-based entity extraction identifies entities using predefined rules such as regular expressions (Regex) and keyword matching.
</p>

<h2><b>TECHNIQUES USED</b></h2>

<p><b>1. REGEX (REGULAR EXPRESSIONS)</b></p>
<ul>
<li>Used to extract patterns like time and date</li>
<li>Example: "3pm", "6pm", "5am"</li>
</ul>

<p><b>2. KEYWORD MATCHING</b></p>
<ul>
<li>Used for detecting tasks, locations, and persons</li>
<li>Example: "meeting", "call", "Chennai", "John"</li>
</ul>

<h2><b>EXAMPLES</b></h2>

<ul>
<li>Time → 3pm, 6pm</li>
<li>Date → tomorrow, Monday</li>
<li>Person → John, Ravi</li>
</ul>

<h2><b>ADVANTAGES</b></h2>

<ul>
<li>Fast and efficient</li>
<li>No training required</li>
<li>Easy to understand and implement</li>
<li>Works well for simple patterns</li>
</ul>

<h2><b>LIMITATIONS</b></h2>

<ul>
<li>Not scalable for complex data</li>
<li>Fails with unseen patterns</li>
<li>Requires manual rule updates</li>
<li>Lower accuracy compared to ML models</li>
</ul>

</div>

```python
import re

def rule_based_ner(text):
    entities = {
        "person": None,
        "date": None,
        "time": None,
        "location": None,
        "task": None
    }

    # TIME (Regex)
    time_match = re.search(r'\b\d{1,2}(am|pm)\b', text.lower())
    if time_match:
        entities["time"] = time_match.group()

    # DATE (Keywords)
    date_keywords = ["tomorrow", "today", "monday", "friday", "december"]
    for word in date_keywords:
        if word in text.lower():
            entities["date"] = word

    # PERSON (Simple Capital Words)
    words = text.split()
    for word in words:
        if word.istitle():
            entities["person"] = word

    # LOCATION (Keywords)
    locations = ["chennai", "delhi", "hotel", "hospital", "apollo"]
    for loc in locations:
        if loc in text.lower():
            entities["location"] = loc

    # TASK (Keywords)
    if "meeting" in text.lower():
        entities["task"] = "schedule meeting"
    elif "call" in text.lower():
        entities["task"] = "make call"
    elif "book" in text.lower():
        entities["task"] = "booking"

    return entities
```

```python
texts = [
    "Schedule meeting with John tomorrow at 3pm",
    "Set reminder to call mom at 6pm",
    "Cancel meeting with Alex",
    "Book a table at Taj Hotel at 8pm",
    "Meeting with Dr. Smith on Monday",
    "Submit report before 25th December",
    "Flight from Chennai to Delhi at 5am",
    "Call Ravi at 7pm",
    "Doctor appointment at Apollo Hospital",
    "Team meeting next Friday at 4pm"
]
```

```python
for text in texts:
    print(text)
    print(rule_based_ner(text))
    print("-" * 50)
```

<h1><b>STEP 3: SPACY NER (NAMED ENTITY RECOGNITION)</b></h1>

<h2><b>OVERVIEW</b></h2>

<p>
spaCy is an advanced Natural Language Processing (NLP) library that provides pre-trained models for tasks like Named Entity Recognition (NER).
</p>

<p>
In this step, we use the pre-trained model <b>en_core_web_sm</b> to extract entities such as PERSON, DATE, TIME, and ORG from text.
</p>

<h2><b>ENTITY LABELS</b></h2>

<ul>
<li>PERSON → Names of people</li>
<li>DATE → Dates or days</li>
<li>TIME → Specific time expressions</li>
<li>ORG → Organizations or institutions</li>
<li>GPE / LOC → Locations (cities, places)</li>
</ul>

<h2><b>HOW SPACY DETECTS ENTITIES</b></h2>

<p>
spaCy uses a pre-trained statistical model based on machine learning. It analyzes:
</p>

<ul>
<li>Word context in a sentence</li>
<li>Part-of-speech tags</li>
<li>Dependency parsing</li>
<li>Word embeddings</li>
</ul>

<p>
Using these features, spaCy predicts entity labels for each word in the text.
</p>

<h2><b>ADVANTAGES</b></h2>

<ul>
<li>Fast and efficient</li>
<li>Easy to use with pre-trained models</li>
<li>No need for large training data</li>
<li>Good performance for general tasks</li>
</ul>

<h2><b>LIMITATIONS</b></h2>

<ul>
<li>May miss domain-specific entities</li>
<li>Less accurate than transformer-based models</li>
<li>Limited customization without retraining</li>
</ul>

</div>

```python
!pip install spacy
!python -m spacy download en_core_web_sm
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def spacy_ner(text):
    doc = nlp(text)

    entities = {
        "person": None,
        "date": None,
        "time": None,
        "location": None,
        "task": text
    }

    for ent in doc.ents:
        if ent.label_ == "PERSON":
            entities["person"] = ent.text
        elif ent.label_ == "DATE":
            entities["date"] = ent.text
        elif ent.label_ == "TIME":
            entities["time"] = ent.text
        elif ent.label_ in ["GPE", "LOC", "ORG"]:
            entities["location"] = ent.text

    return entities
```

```python
for text in texts:
    print(text)
    print(spacy_ner(text))
    print("-" * 50)
```

```python
from transformers import pipeline

ner_pipeline = pipeline("ner", model="dslim/bert-base-NER")

def hf_ner(text):
    results = ner_pipeline(text)

    entities = {
        "person": None,
        "date": None,
        "time": None,
        "location": None,
        "task": text
    }

    for res in results:
        label = res['entity']
        word = res['word']

        if label.startswith("B-PER"):
            entities["person"] = word
        elif label.startswith("B-LOC"):
            entities["location"] = word
        elif label.startswith("B-ORG"):
            entities["location"] = word

    return entities
```

<h1><b>STEP 5: LLM-BASED ENTITY EXTRACTION (SIMULATED)</b></h1>

<h2><b>OVERVIEW</b></h2>

<p>
Due to API cost limitations, a simulated LLM-based approach is implemented.
This method mimics the behavior of a Large Language Model by combining outputs from multiple techniques.
</p>

<h2><b>APPROACH</b></h2>

<ul>
<li>Combine rule-based and spaCy outputs</li>
<li>Fill missing entities intelligently</li>
<li>Return structured JSON output</li>
</ul>

<h2><b>JUSTIFICATION</b></h2>

<p>
This approach simulates how real-world systems integrate multiple models to achieve better performance without relying on expensive APIs.
</p>

</div>

```python
def llm_simulated(text):
    rule = rule_based_ner(text)
    spacy_res = spacy_ner(text)

    final = {}

    for key in rule:
        final[key] = spacy_res[key] if spacy_res[key] else rule[key]

    return final
```

```python
for text in texts:
    print(text)
    print(llm_simulated(text))
    print("-" * 50)
```

```python
import pandas as pd

df = pd.read_csv("intent_dataset_500.csv")

df.head()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
def spacy_extract(text):

    doc = nlp(text)

    result = {
        "person": [],
        "date": [],
        "time": [],
        "location": [],
        "task": text
    }

    for ent in doc.ents:

        if ent.label_ == "PERSON":
            result["person"].append(ent.text)

        elif ent.label_ == "DATE":
            result["date"].append(ent.text)

        elif ent.label_ == "TIME":
            result["time"].append(ent.text)

        elif ent.label_ in ["GPE", "LOC", "ORG"]:
            result["location"].append(ent.text)

    return result
```

```python
ner_model = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)
```

```python
def hf_extract(text):

    output = ner_model(text)

    result = {
        "person": [],
        "date": [],
        "time": [],
        "location": [],
        "confidence": [],
        "task": text
    }

    for item in output:

        label = item["entity_group"]
        word = item["word"]
        score = item["score"]

        if label == "PER":
            result["person"].append(word)
            result["confidence"].append(score)

        elif label in ["LOC", "ORG"]:
            result["location"].append(word)
            result["confidence"].append(score)

    return result
```

```python
def llm_extract(text):

    spacy_res = spacy_extract(text)

    # simple fallback logic
    result = {
        "person": spacy_res["person"],
        "date": spacy_res["date"],
        "time": spacy_res["time"],
        "location": spacy_res["location"],
        "task": text
    }

    return result
```

```python
def clean_format(data):

    return {
        "person": data["person"][0] if data["person"] else None,
        "date": data["date"][0] if data["date"] else None,
        "time": data["time"][0] if data["time"] else None,
        "location": data["location"][0] if data["location"] else None,
        "task": data["task"]
    }
```

```python
import json

def test_input(text):

    print("Input:", text)
    print()

    print("spaCy:")
    print(json.dumps(clean_format(spacy_extract(text)), indent=2))

    print("\nHuggingFace:")
    print(json.dumps(clean_format(hf_extract(text)), indent=2))

    print("\nLLM:")
    print(json.dumps(clean_format(llm_extract(text)), indent=2))
```
